```python
#  Calculating total price
df['TotalPrice'] = df['Quantity'] * df['UnitPrice'] 
```

```python
# In RFM (Recency, Frequency, Monetary) analysis, you need a reference date to calculate 
# Recency — typically the day after the last transaction in your dataset.

# TIMEDELTA
# pd.Timedelta is part of pandas,
# and it's used to represent differences in time — like 1 day, 2 hours, 5 minutes, etc.
# for example, if your latest invoice date is 2023-12-31, this will make snapshot_date = 2024-01-01.

 # Why add 1 day?
# So that no customer ends up having a recency of 0 

Snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days = 1)
Snapshot_date
```

```python
rfm = df.groupby('CustomerID').agg({'InvoiceDate' : lambda x : (Snapshot_date - x.max()).days,
'InvoiceNo' : 'nunique',
'TotalPrice' : 'sum'
})
```

```python
rfm = df.groupby('CustomerID').agg({'InvoiceDate' : lambda x : (Snapshot_date - x.max()).days,
'InvoiceNo' : 'nunique',
'TotalPrice' : 'sum'
})
rfm
```

```python
rfm.rename(columns={'InvoiceDate' : 'Recency',
                     'InvoiceNo': 'Frequency',
                     'TotalPrice' : 'Monetary'},inplace = True)
```

```python
rfm
```

```python
# pd.qcut() is a Pandas function used to divide a numeric column into quantile-based bins
# This splits the Recency column into 5 quantile bins, and assigns:

# 5 to the most recent buyers (lowest values)

# 1 to the least recent buyers (highest values)

rfm['R_score'] = pd.qcut(rfm['Recency'],5,labels = [5,4,3,2,1])
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'),5,labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'],5,labels=[1,2,3,4,5])
```

```python
rfm
```

```python
rfm['RFM_Segment'] =  rfm['R_score'].astype(str) + rfm['F_score'].astype(str) +  rfm['M_Score'].astype(str)
```

```python
rfm
```

```python
rfm['RFM_Score'] = rfm[['R_score','F_score','M_Score']].astype(int).sum(axis=1)
```

```python
rfm
```

```python
def segment_customer(df):
    if df['RFM_Score'] >= 9:
        return 'Champions'
    elif df['RFM_Score'] >= 7:
        return 'Loyal'
    elif df['RFM_Score'] >= 5:
        return 'Potential'
    else:
        return 'At Risk'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
```

```python
rfm
```

```python
def segment_customer(row):
    if row['R_score'] >= 4 and row['F_score'] >= 4:
        return 'Champions'
    elif row['R_score'] >= 3 and row['F_score'] >= 3:
        return 'Loyal Customers'
    elif row['R_score'] >= 4 and row['F_score'] <= 2:
        return 'Potential Loyalist'
    elif row['R_score'] >= 2 and row['F_score'] >= 4:
        return 'Recent Customers'
    elif row['R_score'] <= 2 and row['F_score'] <= 2:
        return 'At Risk'
    else:
        return 'Others'

rfm['Segment'] = rfm.apply(segment_customer, axis=1)
```
